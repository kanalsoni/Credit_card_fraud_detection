```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=[['LR',0.9991456292499094, 0.579310344827586, 0.875],
   [ 'CART',0.9994850368081645, 0.7931034482758621, 0.8914728682170543],
      ['NB',0.9784651756141521,0.8689655172413793, 0.06471494607087827],
     ['SVM',0.9994616293903538, 0.7310344827586207, 0.9380530973451328],
     ['KNN',0.9995201479348805, 0.7862068965517242, 0.9193548387096774],
     ['RFC',0.9995903701883126, 0.8275862068965517, 0.9230769230769231]]
```

```python
df = pd.DataFrame(data, columns = ['Algorithms','Accuracy', 'Precision',  
                                    'Recall'])
```

```python
sns.countplot(x='Algorithms',data=df,palette="gist_earth_r")
plt.show()
```

```python
plt.bar(df['Algorithms'],df['Accuracy']) 
plt.xlabel("Algorithms") 
plt.ylabel("Accuray values") 
plt.title("Comaprison of different algortihms based on their accuracy")
plt.show() 
```

```python
colors=['hotpink','mediumturquoise','seagreen','goldenrod','salmon','olive']
```

```python
plt.bar(df['Algorithms'],df['Accuracy'],color=colors) 
plt.xlabel("Algorithms") 
plt.ylabel("Accuray values") 
plt.title("Comaprison of different algortihms based on their accuracy")
plt.show()
```
